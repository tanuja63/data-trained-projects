```python
from sklearn.preprocessing import StandardScaler
```

```python
ds_new.skew()
```

```python
x_data=ds_new.drop('AveragePrice',axis=1)
```

```python
y_data=ds_new['AveragePrice']
```

```python
x_data.head()
```

```python
y_data.head()
```

```python
for i in x_data:
    x_data[i]=np.log(1+x_data[i])
```

removed skewness from data columns

```python
x_data.skew()
```

```python

for i in x_data.columns:
    sns.distplot(x_data[i])
    plt.figure()
```

 above distplot shows the data is now normally distributed as we remove skewness

```python
st=StandardScaler()
```

```python
x_scaled=st.fit_transform(x_data)
```

```python
x=pd.DataFrame(x_scaled,columns=x_data.columns)
```

scaled data using standard scaler

```python
x.head()
```

Importing regression algorithms

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

splitting data for testing and training

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_data,test_size=.30,random_state=45)
print('x_train shape',x_train.shape)
print('x_test shape',x_test.shape)
print('y_train shape',y_train.shape)
print('y_test shape',y_test.shape)
```

```python
le=LinearRegression()
dtr=DecisionTreeRegressor()
sv=SVR()
knn=KNeighborsRegressor()
rf=RandomForestRegressor()
```

```python
list1=[le,dtr,sv,knn,rf]
for i in list1:
    i.fit(x_train,y_train)
    pred=i.predict(x_test)
    print("accuracy_scores of ",i)
    print("r2_score=",r2_score(y_test,pred))
    print("mean_squared_error=",mean_squared_error(y_test,pred))
    print("mean_absolute_error=",mean_absolute_error(y_test,pred))
```

by obseraving above accuracy matrix we concluded that randomforestclassifier gives lowest error

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in list1:
    cv1=cross_val_score(i,x,y_data,cv=5)
    print('cross values of',i)
    print(cv1)
    print(cv1.mean())
```

# Hyper Parameter Tunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 
            "n_estimators"      : [10,20,30],
            "max_features"      : ["auto", "sqrt", "log2"],
            "min_samples_split" : [2,4,8],
            "bootstrap": [True, False],
            }

grid = GridSearchCV(estimator=rf, param_grid=param_grid, n_jobs=-1, cv=5)

grid.fit(x_train,y_train)
```

```python
grid.best_score_ 
```

```python
grid.best_params_
```

```python
rfr=RandomForestRegressor(bootstrap= False,max_features='sqrt',min_samples_split= 2,n_estimators= 30)
rfr.fit(x_train,y_train)
rpred=rfr.predict(x_test)
cv3=cross_val_score(rfr,x,y_data,cv=5)
print("score",cv3)
print("cross score mean value",cv3.mean())
print('mean squared error',mean_squared_error(rpred,y_test))
print("r2_score",r2_score(rpred,y_test))
```
