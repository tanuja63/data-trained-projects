# handling class imablance problem

```python
from imblearn.over_sampling import SMOTE
```

```python
y=x['fraud_reported']
```

```python
x.drop('fraud_reported',axis=1,inplace=True)
```

```python
sm=SMOTE()
```

```python
x_1,y_1=sm.fit_resample(x,y)
```

# Standardalized data using standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x1=sc.fit_transform(x_1)
```

```python
X1=pd.DataFrame(x1,columns=x_1.columns)
```

```python
X1
```

```python
#import classification algorithmns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#spliting data for training and testing
x_train,x_test,y_train,y_test=train_test_split(X1,y_1,test_size=.20,random_state=47)
print("x_train shape",x_train.shape)
print("x_test shape",x_test.shape)
print("y_train shape",y_train.shape)
print("y_test shape",y_test.shape)
```

```python
lg=LogisticRegression()
sv=SVC()
dtc=DecisionTreeClassifier()
gnb=GaussianNB()
rf=RandomForestClassifier()
ad=AdaBoostClassifier()
```

```python
#fit data into model and predict the values
list1=[lg,sv,dtc,gnb,rf,ad]
for i in list1:
    i.fit(x_train,y_train)
    pred=i.predict(x_test)
    print("accuracy_scores",i)
    print(accuracy_score(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```

by observing above accuracy_scores randomforestclassifier and adaboostclassfier have high accuracy_scores, so lets check cross val scores

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in list1:
    scores=cross_val_score(i,X1,y_1,cv=5)
    print("cross_val_scores",i)
    print(scores)
    print(scores.mean())
    
```

bt observing cross scores we can conclude that randomforestclassifier gives high accuracy
