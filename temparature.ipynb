```python
df['station']=df['station'].fillna(df['station'].mean())
```

```python
#remove outliers using zscore
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df))
```

```python
z
```

```python
print(np.where(z>3))
```

```python
df2=df[(z<3).all(axis=1)]
```

```python
print("shape before removing outliers",df.shape)
print("shape after removing outliers",df2.shape)
```

```python
df2.head()
```

```python
df2.skew()
```

data set have some skewness

```python
df3=df2.iloc[:]
```

```python
#remove skewness usnig cbrt
for i in df3.columns:
    if df3[i].skew()>0.5:
        df3[i]=np.cbrt(df3[i])
    if df3[i].skew()<-0.5:
        df3[i]=np.cbrt(df3[i])
```

```python
df3.skew()
```

```python
#create new dataframe of input features
x=df3.drop(['Next_Tmax','Next_Tmin'],axis=1)
```

```python
#split target variables
y=df3.iloc[:,-2:]
```

```python
y
```

```python
#import regression algorithmns
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split,cross_validate
```

```python
#standalied data using standardscaler
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
```

```python
x1=st.fit_transform(x)
```

```python
x2=pd.DataFrame(x1,columns=x.columns)
```

Split data for training and testing

```python
x1_train,x1_test,y_train,y_test=train_test_split(x2,y,test_size=.30,random_state=47)
```

```python
print("x1_train shape",x1_train.shape)
print("x1_test shape",x1_test.shape)
print("y_train shape",y_train.shape)
print("y_test shape",y_test.shape)
```

```python
lg=LinearRegression()
dtr=DecisionTreeRegressor()
sv=SVR()
knn=KNeighborsRegressor()
rfr=RandomForestRegressor()
```

```python
#fit data into models and predict values
list1=[lg,dtr,knn,rfr]
for i in list1:
    i.fit(x1_train,y_train)
    pred=i.predict(x1_test)
    print("accuracy_scores",i)
    print("r2_score",r2_score(y_test,pred))
    print("mean_squared_error",mean_squared_error(y_test,pred))
    print("mean_absolute_error",mean_absolute_error(y_test,pred))
   
```

by abserving above accuracy matrix random forest gives very less errors

```python
#checking cross val scores
for i in list1:
    
    scores = cross_validate(i, x2, y, cv=5,scoring=('r2', 'neg_mean_squared_error'),return_train_score=True)
    print("cross_validates_scores",i)
    print(scores['test_neg_mean_squared_error'])
    print(scores["train_r2"])
    
```

```python
from sklearn.model_selection import cross_val_score
for i in list1:
    cv1=cross_val_score(i,x2,y,cv=5)
    print("cross_val_scores",i)
    print("scores",cv1)
    print(cv1.mean())
```

# HyperparameterTunning

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 
            "n_estimators"      : [10,20,30],
            "max_features"      : ["auto", "sqrt", "log2"],
            "min_samples_split" : [2,4,8],
            "bootstrap": [True, False],
            }

grid = GridSearchCV(estimator=rfr, param_grid=param_grid, n_jobs=-1, cv=5)

grid.fit(x1_train,y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(bootstrap=False,max_features= 'log2',min_samples_split=4,n_estimators=30)
rf.fit(x1_train,y_train)
rpred=rf.predict(x1_test)
cv3=cross_val_score(rf,x2,y,cv=5)
print("score",cv3)
print("cross score mean value",cv3.mean())
print('mean squared error',mean_squared_error(rpred,y_test))
```

by hyperparametertunning randomforestregressor performs well
